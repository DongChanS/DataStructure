```python
"""
이진트리

1. 이진트리의 노드
2. 이진트리 골격 구현
3. 이진트리 순회 구현
    preorder, levelorder, postorder

"""
class Node:
    def __init__(self):
        self.left = None
        self.right = None
        self.data = None
    def setData(self,data):
        self.data = data
    def setLeft(self,left):
        self.left = left
    def setRight(self,right):
        self.right = right
    def getData(self):
        return self.data
    def getLeft(self):
        return self.left
    def getRight(self):
        return self.right

class BinTree:
    def __init__(self,data):
        self.root = Node()
        self.root.setData(data)
    def append(self,data,direction,base):
        NewNode = Node()
        NewNode.setData(data)
            
        if direction == "->":
            base.setRight(NewNode)
            return base.getRight()
        elif direction == "<-":
            base.setLeft(NewNode)
            return base.getLeft()

def preorder(tree,base):
    print(base.getData(), end=" ")
    if base.getLeft():
        preorder(tree,base.getLeft())
        
    if base.getRight():
        preorder(tree,base.getRight())
        
"""
print(노드1)
preorder(tree,노드2)
    print(노드2)
    preorder(tree,노드4)
        print(노드4)
    preorder(tree,노드5)
        print(노드5)
preorder(tree,노드3)
    print(노드3)

"""



"""
중위순회

level이 높은 순서부터 우선권을 가진다.

        1
    2       3
4      5

4 -> 2 -> 5 -> 1 -> 3

의 순서대로 출력되야함.

"""

def levelorder(tree, base):
    if base.getLeft():
        levelorder(tree,base.getLeft())
    
    print(base.getData(),end=" ")
    
    if base.getRight():
        levelorder(tree,base.getRight())
    
"""
preorder(tree,노드1)
    preorder(tree,노드2)
        preorder(tree,노드4)
            print(4)
        print(2)
        preorder(tree,노드5)
            print(5)
    print(1)
    preorder(tree,노드3)
        print(3)
"""



def postorder(tree, base):
    print(base.getData(),end=" ")
    
    if base.getRight():
        postorder(tree,base.getRight())
        
    if base.getLeft():
        postorder(tree,base.getLeft())
    

```

```python
if __name__ == "__main__":
    tree = BinTree(1)
    second = tree.append(2,"<-",tree.root)
    third = tree.append(3,"->",tree.root)
    fourth = tree.append(4,"<-",second)
    fifth = tree.append(5,"->",second)

    preorder(tree,tree.root)
    levelorder(tree,tree.root)
    postorder(tree,tree.root)
```
